```python
from tensorflow.keras.models import load_model
import pickle

model = load_model("model.keras")
with open("scaler.pkl", "rb") as f:
  scaler = pickle.load(f)
```

```python
import numpy as np
```

```python
# Step 3: Prepare your input sample
# This must be in the same format and shape as training data (e.g., 30 features for breast cancer dataset)
# Replace these numbers with actual feature values
sample_input = np.array([
    17.99, 10.38, 122.8, 1001.0, 0.1184, 0.2776, 0.3001, 0.1471, 0.2419, 0.07871,
    1.095, 0.9053, 8.589, 153.4, 0.006399, 0.04904, 0.05373, 0.01587, 0.03003, 0.006193,
    25.38, 17.33, 184.6, 2019.0, 0.1622, 0.6656, 0.7119, 0.2654, 0.4601, 0.1189
])  # This is a sample from the breast cancer dataset
```

```python
# Step 4: Scale the input
sample_scaled = scaler.transform([sample_input])  # 2D array
```

```python
print(sample_scaled)
```

```python
# Step 5: Make prediction
prediction = model.predict(sample_scaled, verbose=0)
```

```python
# Step 6: Interpret the result
predicted_class = int(prediction[0][0] >= 0.5)
label = "Benign" if predicted_class == 1 else "Malignant"

print(f"Prediction: {label} (probability: {prediction[0][0]:.4f})")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets /Kung fu Panda/KFP1Script.csv", encoding="windows-1252")
df.head()
```

```python
df = df.drop('x', axis=1)
df
```

```python
df.rename(columns={
    '#VALUE!': 'Charecter',
    '#VALUE!.1': 'Statement'
}, inplace=True)
df
```

```python
df['Charecter'].value_counts()
```

```python
df['Statement'].value_counts()
```

```python
df = df[df['Charecter'] != '#VALUE!']
df
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    return text

df['Statement'] = df['Statement'].apply(clean_text)
```

```python
df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import joblib

# Step 1: Load the data
# df = pd.read_csv('df')

# Step 2: Encode labels
le = LabelEncoder()
df['character_encoded'] = le.fit_transform(df['Charecter'])

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(df['Statement'], df['character_encoded'], test_size=0.2, random_state=42)

# Step 4: Build pipeline
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

# Step 5: Train the model
model.fit(X_train, y_train)

# Step 6: Save model and label encoder
joblib.dump(model, 'kungfu_ml_model.pkl')
joblib.dump(le, 'kungfu_label_encoder.pkl')
```

```python
# Load model and encoder
import joblib

model = joblib.load('kungfu_ml_model.pkl')
le = joblib.load('kungfu_label_encoder.pkl')

# Predict function
def predict_character(dialogue):
    encoded = model.predict([dialogue])[0]
    return le.inverse_transform([encoded])[0]

# Example
print(predict_character("relieved oh so um i should stop talking"))
```
